## continues sampling

```python
from model import *
from Mess_60Ghz.dataset import *

dataset = WifiCSIDataset('Mess_60Ghz/data/60Ghz_position_data.csv', 5, continue_sampling=False)
default_test = TestModel("60Ghz_sample_size_test",dataset, batch_size=20)
training_size, validation_size, test_size = default_test.get_dataset_count()
print(f"training size: {training_size} validation_size: {validation_size} test size: {test_size}")
default_test.train()
default_test.test()

dataset = WifiCSIDataset('Mess_60Ghz/data/60Ghz_position_data.csv', 20, continue_sampling=True, recreate=True)
default_test = TestModel("60Ghz_continous_samling",dataset, batch_size=100)
training_size, validation_size, test_size = default_test.get_dataset_count()
print(f"training size: {training_size} validation_size: {validation_size} test size: {test_size}")
default_test.train(True)
default_test.test(True)
```
